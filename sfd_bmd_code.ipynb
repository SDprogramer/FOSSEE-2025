```python
import pandas as pd
import numpy as np
```

```python
file_path = "C:\\ZZZ_PROGRAMING\\Nit-DGP\\FOSSEE 2025\\SFD_BMD_Data.xlsx"
```

```python
xls = pd.ExcelFile(file_path)
```

```python
xls.sheet_names
```

```python
df = pd.read_excel(file_path, sheet_name = "Sheet1")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
distance = df["Distance (m)"]
shear_force = df["SF (kN)"]
bending_moment = df["BM (kN-m)"]
```

```python
fig, axes = plt.subplots(2, 1, figsize = (10, 8), sharex = True)
```

```python
axes[0].plot(distance, shear_force, 'bo-', label = "Shear Force")
axes[0].axhline(0, color = "black", linewidth = 1)
axes[0].set_ylabel("Shear Force (kN)")
axes[0].set_title("Shear Force Diagram (SFD)")
axes[0].grid(True)
axes[0].legend()
```

```python
axes[1].plot(distance, bending_moment, 'rs-', label = "Bending Moment")
axes[1].axhline(0, color = "black", linewidth = 1)
axes[1].set_xlabel("Distance (m)")
axes[1].set_ylabel("Bending Moment (kN-m)")
axes[1].set_title("Bending Moment Diagram (BMD)")
axes[1].grid(True)
axes[1].legend()
```

```python
plt.tight_layout()
```
